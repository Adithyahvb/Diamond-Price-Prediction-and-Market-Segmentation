```python
#CHECKING FOR OUTLIERS
plt.figure(figsize = (10, 6))
plt.suptitle("Boxplots for numerical columns to check outliers", fontsize=18, fontweight='bold')
for i, col in enumerate(numcols, 1):
    plt.subplot(2, 4, i)
    sns.boxplot(y= data[col], color = 'skyblue')
    plt.title(f'{col}', fontweight = 'bold')
plt.tight_layout()
plt.show()
```

```python
#CAPPING THE OUTLIERS
df = data.copy()
for col in numcols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower_bound,upper_bound)
```

```python
#OUTLIERS ARE HANDLED
plt.figure(figsize = (10,6))
plt.suptitle("Box plot after handling outliers")

for i, col in enumerate(numcols, 1):
    plt.subplot(2,4,i)
    sns.boxplot(y = df[col], color = 'skyblue')
    plt.title(f'{col}', fontweight = 'bold')
plt.tight_layout()
plt.show()
```

```python
#CHECKING FOR SKEWNESS
for i in numcols:
    print(f' {i} {data[i].skew()}')
```

```python
#SKEWNESS HANDLED USING LOG1P TRANSFORM
for i in numcols:
    skew = df[i].skew()
    print(f'{i} : {skew}')
    if skew > 0.8:
        df[i] = np.log1p(df[i])
    print(f'After handling skewness the value is {df[i].skew()}')
```

```python
#ORDINAL ENCODING- MANUAL MAPPING
cut_map = {'Fair': 1, 'Good': 2, 'Very Good': 3, 'Premium': 4, 'Ideal': 5}

color_map = {'J': 1, 'I': 2, 'H': 3, 'G': 4, 'F': 5, 'E': 6, 'D': 7 }

clarity_map = {'I1': 1, 'SI2': 2, 'SI1': 3, 'VS2': 4, 'VS1': 5, 'VVS2': 6, 'VVS1': 7, 'IF': 8 }

df['cut_encoded'] = df['cut'].map(cut_map)
df['color_encoded'] = df['color'].map(color_map)
df['clarity_encoded'] = df['clarity'].map(clarity_map)
```

```python
#DERIVING NEW FEATURES
df['volume'] = df['x'] * df['y'] * df['z']
df['price_per_carat'] = df['price'] / df['carat']
df['dimension_ratio'] = (df['x'] + df['y']) / (2*df['z'])
def carat_cat(carat):
    if carat < 0.5:return 'Light'
    elif carat <= 1.5: return 'Medium'
    else: return 'Heavy'
df['carat_category'] = df['carat'].apply(carat_cat)
df['size_index'] = (df['x'] + df['y'] + df['z']) / 3
df['quality_score'] = df['cut_encoded'] + df['color_encoded'] + df['clarity_encoded']
df['density'] = df['carat'] / df['volume']
df['symmetry'] = abs(df['x'] - df['y'])
df['log_carat'] = np.log1p(df['carat'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
```

```python
models = {"Linear Regression": LinearRegression(),
         "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
         "Decision Tree": DecisionTreeRegressor(random_state = 42),
         "KNN": KNeighborsRegressor(n_neighbors=5),
         "XGBoost": XGBRegressor(random_state = 42)}
```

```python
model_results = {}
trained_models = {}
for name, model in models.items():
    print(f'"Model name" {name}')
    
    model.fit(x_train, y_train)
    
    y_pred = model.predict(x_test)
    
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    model_results[name] = { 'MAE' : round(mae,4), 'MSE' : round(mse,4), 'RMSE' : round(rmse,4), 'R2' : round(r2,4)}
    
    trained_models[name] = model
    
    print(f' MAE : {mae:.4f}')
    print(f' MSE : {mse:.4f}')
    print(f' RMSE : {rmse:.4f}')
    print(f' R2 : {r2:.4f}')
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import StandardScaler
import numpy as np

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)


x_train_np = x_train_scaled.astype(np.float32)
x_test_np = x_test_scaled.astype(np.float32)

y_train_np = y_train.to_numpy(dtype=np.float32).reshape(-1, 1)  
y_test_np = y_test.to_numpy(dtype=np.float32).reshape(-1, 1)

#converting to PyTorch tensors
x_train_tensor = torch.FloatTensor(x_train_np)
x_test_tensor = torch.FloatTensor(x_test_np)
y_train_tensor = torch.FloatTensor(y_train_np)
y_test_tensor = torch.FloatTensor(y_test_np)

class SimpleANN(nn.Module):
    #constructor
    def __init__(self, input_size):
        super(SimpleANN, self).__init__()#super constructor

        #Hidder layers
        self.fc1 = nn.Linear(input_size, 32)
        self.fc2 = nn.Linear(32,64)
        self.fc3 = nn.Linear(64,32)
        self.fc4 = nn.Linear(32,16)
        
        self.output = nn.Linear(16,1)
        
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        #x = self.relu(self.fc2(x))
        #x = self.fc3(x)
        x = self.relu(self.fc3(x))
        x = self.fc4(x)
        #x = self.relu(self.fc4(x))
        x = self.output(x)
        return x

ann_model = SimpleANN(x_train.shape[1])

criterion = nn.MSELoss()
optimizer = optim.Adam(ann_model.parameters(),lr=0.001)

epochs = 50 

for epoch in range(epochs):
    outputs = ann_model(x_train_tensor)
    loss = criterion(outputs, y_train_tensor)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch+1) % 10 ==0:
        print(f' Epoch[{epoch+1}/{epochs}],Loss : {loss.item():.6f}')

with torch.no_grad():
    ann_model.eval()
    y_pred_ann = ann_model(x_test_tensor).numpy()

mae_ann = mean_absolute_error(y_test, y_pred_ann)
mse_ann = mean_squared_error(y_test, y_pred_ann)
rmse_ann = np.sqrt(mse_ann)
r2_ann = r2_score(y_test, y_pred_ann)

print(f'\n ANN Results')
print(f' MAE: {mae_ann:.4f}')
print(f' MSE: {mse_ann:.4f}')
print(f' RMSE: {rmse_ann:.4f}')
print(f' r2: {r2_ann:.4f}')

model_results['ANN'] = { 'MAE' : round(mae_ann,4), 'MSE' : round(mse_ann,4), 'RMSE' : round(rmse_ann,4), 'R2' : round(r2_ann,4) }
trained_models['ANN'] = ann_model 
    
```

```python
#COMPARING AND SAVING THE BEST PERFORMING MODEL
import joblib
results_df = pd.DataFrame(model_results).T.sort_values('R2', ascending = False)
print(results_df.round(4))

best_model_name = results_df['R2'].idxmax()
best_score = results_df.loc[best_model_name, 'R2']

print(f'\n Best performing model is {best_model_name}')
print(f' R2 score : {best_score:.4f}')

best_model = trained_models[best_model_name]

# joblib.dump(best_model, 'Diamond_price_prediction_best_model.joblib')
# print("Best performing model is saved")
```

```python
#IMPORTING PACKAGES FOR KMEANS CLUSTERING
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
k_df = pd.read_csv("Diamond_project_complete_preprocessed.csv")
```

```python
#DERIVING THE LOG PRICE FEATURE 
k_df['log_price'] = np.log1p(k_df['price'])
```

```python
k_df.columns
```

```python
#FINAL FEATURES FOR CLUSTERING
k_x = k_df[['log_price','price_per_carat','log_carat','quality_score']].copy()
k_x
```

```python
k_x.columns
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(k_x)#SCALING THE DATA
inertia = []
silhouette_scores = []

k_range = range(2,11)
#RUNNING K-MEANS FOR DETERMINING THE BEST POSSIBLE CLUSTERS
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(x_scaled)

    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(x_scaled, labels))
```
